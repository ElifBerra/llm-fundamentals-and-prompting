```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv(override=True)
api_key = os.getenv('GROQ_API_KEY')

# OpenAI kütüphanesini Groq altyapısına bağlıyoruz
client = OpenAI(
    api_key=api_key,
    base_url="https://api.groq.com/openai/v1"
)

resp = client.chat.completions.create(
    model='llama-3.3-70b-versatile',  # Groq üzerindeki en güçlü ve ücretsiz model
    messages=[
        {'role': 'system', 'content': 'Sen yardımcı bir asistansın. Kısa cevap ver.'},
        {'role': 'user', 'content': 'Overfitting nedir?'},
    ],
)

print("=== CEVAP ===")
print(resp.choices[0].message.content)

print("\n=== TOKEN KULLANIMI ===")
print(resp.usage)                     # kaç token harcandı?

print("\n=== BİTİŞ SEBEBİ ===")
print(resp.choices[0].finish_reason)
```

```python
# Problem: "Bir mağazada 3 kutu var, her kutuda 12 kalem. 5 kalem satıldı. Kaç kalem kaldı?"

# 1. DOĞRUDAN CEVAP (Sadece cevabı yaz)
direct_prompt = """Bir mağazada 3 kutu var, her kutuda 12 kalem. 5 kalem satıldı. Kaç kalem kaldı?
SADECE nihayi sonucu sayı olarak yaz, hiçbir açıklama yapma."""

# 2. CHAIN-OF-THOUGHT (Adım adım düşün)
cot_prompt = """Bir mağazada 3 kutu var, her kutuda 12 kalem. 5 kalem satıldı. Kaç kalem kaldı?
Adım adım düşün, hesaplamalarını açıkla ve en sonda cevabı ver."""

# Doğrudan Çağrı
resp_direct = client.chat.completions.create(
    model='llama-3.3-70b-versatile',
    messages=[{'role': 'user', 'content': direct_prompt}]
)

# CoT Çağrısı
resp_cot = client.chat.completions.create(
    model='llama-3.3-70b-versatile',
    messages=[{'role': 'user', 'content': cot_prompt}]
)

print("=== ⚡ DOĞRUDAN YANIT (Sadece Cevap) ===")
print(resp_direct.choices[0].message.content)

print("\n" + "="*50 + "\n")

print("=== 🧠 CHAIN-OF-THOUGHT (Adım Adım) ===")
print(resp_cot.choices[0].message.content)
```

```python
import json

# JSON Prompt (Görseldeki örneğe birebir uygun)
json_prompt = """Aşağıdaki metinden bilgi çıkar ve SADECE geçerli JSON döndür.
Şema: {"kisi": str, "sehir": str, "yas": int}
Açıklama veya markdown ekleme.

Metin: Elif Ankara'da yaşayan 21 yaşında bir mühendislik öğrencisi."""

# API Çağrısı
resp_json = client.chat.completions.create(
    model='llama-3.3-70b-versatile',
    messages=[{'role': 'user', 'content': json_prompt}]
)

raw_output = resp_json.choices[0].message.content

print("=== 📥 MODELİN DÖNDÜRDÜĞÜ HAM (RAW) METİN ===")
print(raw_output)
print("\n" + "="*50 + "\n")

# 1. DOĞRUDAN PARSE ETME DENEMESİ
try:
    data = json.loads(raw_output)
    print("✅ Doğrudan json.loads() BAŞARILI!")
    print("Parse Edilen Veri:", data)
    print("Kişi:", data.get("kisi"))
except json.JSONDecodeError as e:
    print("❌ Doğrudan json.loads() HATA ALDI!")
    print("Hata Nedeni:", e)
```
